# Pre & post covid-19 analysis of software listings & their reviews on capterra.com

Ever since the pandemic the landscape of office work has been forced to change around the world, this has led to emergence  and decline of many  technologies. A lot of office productivity apps has seen an explosion on play store and It will be interesting to see the number of softwares and their reviews before March 2020 and afterwards to capture post covid and precovid era on capterra.com, an online marketplace vendor serving as an intermediary between buyers and sellers within the software industry.


We start with scraping the data off the capterra site. We will scrape the info about the products and other relevant details and the product reviews.

# Scraping the sitemap with BeautifulSoup

We start with scraping the link to each categorie from the sitemap,which is in XML, using BeautifulSoup.

We use user agent header as without it, it displays Err : 403(Forbidden), beacuse the site detects that a bot is trying to access the page.

We then save it as a csv file and load it into a pandas dataframe.

# Dynamic scraping using selenium

BeautifulSoup is quite a powerful library in itself, but it falls short when we are scraping a dynamic JavaScript site like capterra, where certain parts of the page are rendered by JavaScript links. Here we use selenium to imitate a real browser and click "Show More" button on the categories site and other buttons in the product description as well. 

In the code snippet above we are using the click method of the webdriver class it repeatedly to render all the products links in a particular category on a single page, which we then scrape using BeautifulSoup. We again save this as csv file and load it as a pandas dataframe

Here above we have 787 software listings related to data, of which we'll scrape reviews.  

# Scraping the product info and reviews 

We now loop through the links from the dataframe and use selenium to load the main page of the product from we where scrape info about the product, what's it best suitable for, its contact details and other features and specifications with BeautifulSoup. And the in the same loop load another page of reviews where we click "Show More Reviews" to load all the reviews in a single and then scrape them, same as we did with the product listings. We then load each row for each listing and columns for various extracted fields.
